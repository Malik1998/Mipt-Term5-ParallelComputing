```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import pandas as pd
from statsmodels.nonparametric.kde import kernel_switch, KDEUnivariate
from statsmodels.distributions.empirical_distribution import ECDF

%matplotlib inline
```

 A = 100 x 100   
 B = 100 x 100

```python
ns = np.array([1, 2, 3, 4])
time = np.array([0.03, 16.54, 12.99, 9.33])
```

```python
plt.figure(figsize=(10, 10))
plt.plot(ns, time);
```

A = 1000 x 100                        
B = 100 x 1000

```python
ns = np.append(np.arange(1, 10), [20])
time = np.array([2.25, 1.16, 0.82, 0.6, 0.43, 0.38, 0.37, 0.27, 0.2, 0.11])
plt.figure(figsize=(10, 10))
plt.plot(ns, time);
```

**Ускорение**

Ускорением параллельного алгоритма называют отношение времени выполнения
лучшего последовательного алгоритмам к времени выполнения параллельного алгоритм

```python
plt.figure(figsize=(10, 10))
plt.plot(ns, time[0] / time);
```

** Эффективность **

Эффективность равна отношению ускорения на количество потоков

```python
plt.figure(figsize=(10, 10))
plt.plot(ns, (time[0] / time) / ns);
```
